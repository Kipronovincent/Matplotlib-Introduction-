```python
import pandas as pd 
import numpy as np 
import matplotlib.pyplot as plt
```

```python
# Plotting x and y points 
xpoints=np.array([0,7])
ypoints=np.array([10,100])
plt.plot(xpoints,ypoints)
plt.show()
```

```python
#plotting without a line 
xpoints=np.array([2,10])
ypoints=np.array([20,30])
plt.plot(xpoints,ypoints,"*")
plt.show()
```

```python
#plotting multiple  points
xpoints=np.array([1,2,3,4,5])
ypoints=np.array([10,15,7,9,30])
plt.plot(xpoints,ypoints)
plt.show()
```

```python
#Default X points
xpoints= np.array([8,2,9,30,5])
plt.plot(xpoints)
plt.show()
```

# Markers 


```python
# "o" string markers . used to represent each data point with a cirular marker 
ypoints=np.array([1,3,8,46])
plt.plot(ypoints,marker="o")
plt.show()
```

```python
#A astar can also be used as a marker 
ypoints=np.array([2,3,4,9,15])
plt.plot(ypoints,marker="*")
plt.show()
```

# 2. format strings fmt

```python
# fmt is written with syntax ; marker|line|color e.g "o:r"  . Here "o" represents the marker style , ":"shows the line style ,"r"
#represents the color of the line and the marker 

xpoints= np.array([3,9,3,4])
ypoints= np.array([4,6,9,12])
plt.plot(xpoints,ypoints,"*:y")
plt.show()
```

# Marker Size 

```python
# Marker Size(ms) is used to set the size of the marker
ypoints=np.array([1,3,8,46])
plt.plot(ypoints,marker="o", ms=5)
plt.show()
```

# Line Styles 

```python
# There are different line styles i.e dashed (--), dotted(:), solid(-), dashdot(-.)
xpoints=np.array([0,7])
ypoints=np.array([10,100])
plt.plot(xpoints,ypoints, ls= "-.")
plt.show()
```

Line color 

```python
# color(c)  is  used to is used to set the color of the line
ypoints=np.array([1,3,8,4,30])
plt.plot(ypoints, c="y")
plt.show()
```

Multiple lines 

```python
#Many lines can be plotted by adding more plt.plot() functions
x1=np.array([2,3,5,6])
x2=np.array([3,5,9,11])
plt.plot(x1)
plt.plot(x2)
plt.show()
```

```python
x1 = np.array([0, 1, 2, 3])
y1 = np.array([3, 8, 1, 10])
x2 = np.array([0, 1, 2, 3])
y2 = np.array([6, 2, 7, 11])
plt.plot(x1, y1, x2, y2)
plt.show()
```

# creating labels for a plot

```python
# xlabel and ylabel functions are used to set the labels for the x and y axis 
x=np.array([1,2,3,4,5,6])
y=np.array([5,15,80,100,30,50])
plt.plot(x,y)
plt.xlabel("Month")
plt.ylabel("Amount of rainfall in millimeters")
plt.show()
```

creating a title for a plot

```python
# title() is used to set a title for a plot
x=np.array([1,2,3,4,5,6])
y=np.array([5,15,80,100,30,50])
plt.plot(x,y)
plt.xlabel("Month")
plt.ylabel("Amount of rainfall in millimeters")
plt.title("Rainfall Data")
plt.show()
```

setting the font properties for the title and labels

```python
# fontdict is used to set the font properties 
x=np.array([1,2,3,4,5,6])
y=np.array([5,15,80,100,30,50])
font1={"family":"serif","color":"blue","size":25}
font2={"family":"Arial","color":"green","size":15}
font3={"family":"Times New Roman","color":"darkred","size":10}
plt.plot(x,y)
plt.xlabel("Month",fontdict=font2)
plt.ylabel("Amount of rainfall in millimeters",fontdict=font3)
plt.title("Rainfall Data",fontdict=font1)
plt.show()
```

Adding grids to a plot 

```python
# grid() function is used to add grid lines to a plot 
x=np.array([1,2,3,4,5,6])
y=np.array([5,15,80,100,30,50])
font1={"family":"serif","color":"blue","size":25}
font2={"family":"Arial","color":"green","size":15}
font3={"family":"Times New Roman","color":"darkred","size":10}
plt.plot(x,y)
plt.xlabel("Month",fontdict=font2)
plt.ylabel("Amount of rainfall in millimeters",fontdict=font3)
plt.title("Rainfall Data",fontdict=font1)
plt.grid()
plt.show()
```

```python
# when you  want to specify lines to display 
x=np.array([1,2,3,4,5,6])
y=np.array([5,15,80,100,30,50])
font1={"family":"serif","color":"blue","size":25}
font2={"family":"Arial","color":"green","size":15}
font3={"family":"Times New Roman","color":"darkred","size":10}
plt.plot(x,y)
plt.xlabel("Month",fontdict=font2)
plt.ylabel("Amount of rainfall in millimeters",fontdict=font3)
plt.title("Rainfall Data",fontdict=font1)
plt.grid(axis="y")
plt.show()
```
